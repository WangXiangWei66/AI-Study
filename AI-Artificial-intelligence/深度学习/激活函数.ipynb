```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 为什么神经网络要有激活函数
# 激活函数可以让神经网络处理非线性问题. 
```

```python
# 最最常用的激活函数就是relu
def relu(x):
    if x > 0:
        return x
    else:
        return 0
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, [relu(i) for i in x])
```

```python
# 带泄露的relu  leaky relu 小于0部分也有导数. 但是事实证明, 这种激活函数, 效果没有比relu更好. 
```

```python
# relu6
def relu6(x):
    if x >= 6:
        return 6
    elif x <= 0:
        return 0
    else:
        return x
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, [relu6(i) for i in x])
```

```python
# sigmoid激活函数容易梯度消失梯度爆炸. 如果是二分类问题, 一般放在最后一层, 对输出进行处理. 
# sigmoid一般不做隐层的激活函数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, [sigmoid(i) for i in x])
```

```python
# tanh  双曲正切激活函数. 在自然语言处理中的某些算法中会使用tanh. 可解释性. 
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, [tanh(i) for i in x])
```

```python
# softplus  # 可以让你传入的值变成正数. 
def softplus(x):
    return np.log(np.exp(x) + 1)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, [softplus(i) for i in x])
plt.plot(x, [relu(i) for i in x], c='r')
```
