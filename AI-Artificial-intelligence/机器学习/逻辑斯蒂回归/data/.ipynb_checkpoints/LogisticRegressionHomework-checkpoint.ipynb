# 0.0.1 手动实现逻辑斯蒂回归梯度下降

```python
#定义sigmod函数
def sigmod(z):
    return 1/(1 + np.exp(-z))
```

```python
#绘制函数图形
nums = np.arange(-8,8,step=1)
fig,ax = plt.subplots(figsize=(10,3))
ax.plot(nums,sigmod(nums),'r')
```

```python
#定义模型
def model(X,theta):
    return sigmod(np.dot(X,theta.T))
```

```python
#定义损失函数
def cost(X,y,theta):
    left = np.multiply(-y,np.log(model(X,theta)))
    right = np.multiply(1-y,np.log(1-model(X,theta)))
    return np.sum(left-right) / len(X)
```

```python
#定义梯度
def gradient(X,y,theta):
    grad = np.zeros(theta.shape)
    error = (model(X,theta)-y).ravel() #计算误差，并展平为一维数组
    for j in range(len(theta.ravel())):
        term = np.multiply(error,X[:,j])
        grad[0,j] = np.sum(term ) / len(X)

    return grad
```

```python
STOP_ITER = 0
STOP_COST = 1 #损失值的变化
STOP_GRAD = 2

def stopCriterion(type,value,threshold):
    if type == STOP_ITER:
        return value > threshold
    elif type == STOP_COST:
        return abs(value[-1]-value[-2]) < threshold #最近两次的损失值之差
    elif type == STOP_GRAD:
        return np.linalg.norm(value) < threshold  #梯度范数停止

def shuffleData(data):
    np.random.shuffle(data)
    cols = data.shape[1]  #获取数据的列数
    X = data[:,0:cols-1]
    y = data[:,cols-1:]
    return X,y
```

```python
#定义梯度下降函数
import time
def descent(data, theta, batchSize, stopType, thresh, alpha):
    init_time = time.time()
    i = 0 
    k = 0 
    X, y = shuffleData(data)
    grad = np.zeros(theta.shape) 
    costs = [cost(X, y, theta)] 

    
    while True:
        grad = gradient(X[k:k+batchSize], y[k:k+batchSize], theta)
        k += batchSize 
        #超出范围就重新打乱顺序
        if k >= n: 
            k = 0 
            X, y = shuffleData(data) 
        theta = theta - alpha*grad 
        costs.append(cost(X, y, theta)) 
        i += 1 

        if stopType == STOP_ITER:       
            value = i
        elif stopType == STOP_COST:     
            value = costs
        elif stopType == STOP_GRAD:     
            value = grad
        if stopCriterion(stopType, value, thresh): break
    
    return theta, i-1, costs, grad, time.time() - init_time
```

```python
#执行优化实验并执行可视化结果
def runExpe(data, theta, batchSize, stopType, thresh, alpha):
    #import pdb; pdb.set_trace();
    theta, iter, costs, grad, dur = descent(data, theta, batchSize, stopType, thresh, alpha)
    #记录是原始数据还是缩放数据
    name = "Original" if (data[:,1]>2).sum() > 1 else "Scaled"
    name += " data - learning rate: {} - ".format(alpha)
    #判断是全批量还是小批量
    if batchSize==n: strDescType = "Gradient"
    elif batchSize==1:  strDescType = "Stochastic"
    else: strDescType = "Mini-batch ({})".format(batchSize)
    name += strDescType + " descent - Stop: "
    #描述一下停止的条件
    if stopType == STOP_ITER: strStop = "{} iterations".format(thresh)
    elif stopType == STOP_COST: strStop = "costs change < {}".format(thresh)
    else: strStop = "gradient norm < {}".format(thresh)
    name += strStop
    print ("***{}\nTheta: {} - Iter: {} - Last cost: {:03.2f} - Duration: {:03.2f}s".format(
        name, theta, iter, costs[-1], dur))
    fig, ax = plt.subplots(figsize=(12,4))
    ax.plot(np.arange(len(costs)), costs, 'r')
    ax.set_xlabel('Iterations')
    ax.set_ylabel('Cost')
    ax.set_title(name.upper() + ' - Error vs. Iteration')
    return theta
```

```python
n=100
runExpe(orig_data, theta, n, STOP_ITER, thresh=5000, alpha=0.000001)
```

```python
#简易版梯度下降
def descent_1(X,y,theta,thresh,alpha):
    i = 0
    grad = np.zeros(theta.shape)
    costs = [cost(X,y,theta)]

    while True:
        grad = gradient(X,y,theta)
        theta = theta-alpha*grad
        costs.append(cost(X,y,theta))
        i+=1

        if i > thresh:
            break

    return theta,costs,grad
```

# 随机梯度下降和小批量梯度下降

```python
from sklearn.datasets import make_blobs
```

```python
data,target = make_blobs(centers=2,random_state = 2048) #random_state确保每次运行生成相同的数据
plt.scatter(data[:,0],data[:,1],c = target)
```

```python
a3 = plt.subplot(111,projection='3d')
a3.scatter(data[:,0],data[:,1],target,c=target)
a3.view_init(30,-60) #设置3D图的视角
```

```python
X = np.hstack((np.ones(shape=(100,1)),data))
```

# 批量梯度下降

```python
#预测函数:线性函数+激活函数
def sigmod(theta):
    z = np.dot(X,theta)
    return 1 / (1 + np.exp(-z))
#损失函数:预测值和真实值的差距
def loss(y_true,y_pred):
    return -np.sum(y_true * np.log(y_pred) + (1-y_true) * np.log(1-y_pred)) / 100
#梯度下降
theta = np.random.randn(3) * 0.01
learning_rate = 0.01
loss_=0
last_loss=1
i=0
while True:
    y_proba = sigmod(theta)
    dtheta = np.dot((y_proba-target),X)/100 #计算参数的梯度
    theta = theta - learning_rate * dtheta

    last_loss = loss_
    loss_ = loss(target,y_proba)

    if np.abs(loss_ - last_loss) < 0.00001:
        break

    i+=1
    if i% 100 == 0:
        y_pred = (y_proba >= 0.5)
        acc = (y_pred == target).mean()
        print(f'第{i}次训练，准确率：{acc},损失:{loss_}')
```

```python
#使用sigmod函数的单神经元模型
def formard_propagation(x,y):
    z = theta[0]*1 + theta[1]*x+theta[2]*y
    a = 1 / (1  + np.exp(-z))
    return a
```

# 随机梯度下降

```python
def sigmoid(X_i,theta):
    z = np.dot(X_i,theta)
    return 1 / (1 + np.exp(-z))

def loss(y_true,y_pred):
    return -np.sum(y_true*np.log(y_pred) + (1-y_true)*np.log(1-y_pred))

theta = np.random.randn(3) * 0.01
learning_rate = 0.01

loss_=0
last_loss = 1
losses=[] #存储训练过程中的损失值
while True:
    for i in range(100):
        X_i = X[i]
        y_i = target[i]
        y_proba = sigmoid(X_i,theta)
        dtheta = np.dot((y_proba-y_i),X_i)
        theta = theta - learning_rate * dtheta

        y_proba = sigmoid(X_i,theta)
        last_loss = loss_
        loss_ = loss(y_i,y_proba)
        losses.append(loss_)
    if np.abs(loss_ - last_loss) < 0.0001:
        break
```

```python
y_proba = sigmoid(X,theta)
y_perd = (y_proba>=0.5)
acc = (y_perd == target).mean()
print(f'预测的准确率：{acc}')
```

```python
def sigmoid(X,theta):
    z = np.dot(X,theta)
    return 1 / (1 + np.exp(-z))

def loss(y_true,y_pred):
    return -np.sum(y_true*np.log(y_pred) + (1-y_true) * np.log(y_pred)) / 100
theta = np.random.randn(3)*0.01
learning_rate = 0.01
loss_=0
last_loss = 1
i=0
losses=[]

while True:
    for j in range(1,11):
        index = np.arange(0,100)
        np.random.shuffle(index)
        #获取批次大小选取训练数据
        X_batch = X[index][:j*10]
        y_batch = target[index][:j*10]

        y_proba = sigmoid(X_batch,theta)
        dtheta = np.dot((y_proba-y_batch),X_batch)/10
        theta = theta-learning_rate*dtheta

        last_loss = loss_
        loss_=loss(y_batch,y_proba)
        losses.append(loss_)

    if np.abs(loss_ - last_loss) < 0.0001:
        break
```
