## 1.2 The logistic Regression

目标：建立分类器（求解出三个参数 $\theta_0         \theta_1         \theta_2 $）


设定阈值，根据阈值判断录取结果

### 要完成的模块
-  `sigmoid` : 映射到概率的函数

-  `model` : 返回预测结果值

-  `cost` : 根据参数计算损失

-  `gradient` : 计算每个参数的梯度方向

-  `descent` : 进行参数更新

-  `accuracy`: 计算精度

###  `sigmoid` 函数

$$
g(z) = \frac{1}{1+e^{-z}}   
$$

```python
def sigmoid(z):
    return 1 / ( 1 + np.exp(-z))
```

```python
nums = np.arange(-10,10,step=1)
fig,ax = plt.subplots(figsize=(12,4))
ax.plot(nums,sigmoid(nums),'r')
```

### Sigmoid
* $g:\mathbb{R} \to [0,1]$
* $g(0)=0.5$
* $g(- \infty)=0$
* $g(+ \infty)=1$

```python
def model(X,theta):
    return sigmoid(np.dot(X,theta.T))
```

$$
\begin{array}{ccc}
\begin{pmatrix}\theta_{0} & \theta_{1} & \theta_{2}\end{pmatrix} & \times & \begin{pmatrix}1\\
x_{1}\\
x_{2}
\end{pmatrix}\end{array}=\theta_{0}+\theta_{1}x_{1}+\theta_{2}x_{2}
$$

```python
if 'Ones' in pdData.columns:
    pdData.drop('Ones', axis=1, inplace=True)
# pdData.insert(0,'Ones',1)

orig_data=pdData.to_numpy() #将DataFrame转化为numpy数组
cols=orig_data.shape[1] #获取列数
X = orig_data[:,0:cols-1]
y=orig_data[:,cols-1:cols]
theta=np.zeros([1,3]) #初始化参数为1行3列的全零数组
```

### 损失函数
将对数似然函数去负号

$$
D(h_\theta(x), y) = -y\log(h_\theta(x)) - (1-y)\log(1-h_\theta(x))
$$
求平均损失
$$
J(\theta)=\frac{1}{n}\sum_{i=1}^{n} D(h_\theta(x_i), y_i)
$$

```python
def cost(X,y,theta):
    left = np.multiply(-y,np.log(model(X,theta)))
    right = np.multiply(1-y,np.log(1-model(X,theta)))
    return np.sum(left-right) / (len(X))
```

```python
cost(X,y,theta)
```

### 计算梯度


$$
\frac{\partial J}{\partial \theta_j}=-\frac{1}{m}\sum_{i=1}^n (y_i - h_\theta (x_i))x_{ij}
$$


```python
def gradient(X,y,theta):
    grad = np.zeros(theta.shape)
    error = (model(X,theta)-y).ravel() #将误差数组从二维展开为一维
    for j in range(len(theta.ravel())):
        term = np.multiply(error,X[:,j])
        grad[0,j]=np.sum(term) / len(X)

    return grad
```

## 1.2.5 Gradint descent

比较3中不同梯度下降方法

```python
STOP_ITER=0 #按照迭代次数停止
STOP_COST=1 # 按代价函数变化量停止
STOP_GRAD=2 # 按梯度停止

def stopCriterion(type,value,threshold):
    if type == STOP_ITER : return value > threshold
    elif type == STOP_COST: return abs(value[-1]-value[-2]) < threshold
    elif type == STOP_GRAD : return np.linalg.norm(value) < threshold    
```

```python
import numpy.random
def shuffleData(data):
    np.random.shuffle(data)
    cols = data.shape[1]
    X = data[:,0:cols-1]
    y = data[:,cols-1:]
    return X,y
```

```python
import time
#batchSize：梯度下降方式
#stopType：停止策略
def descent(data,theta,batchSize,stopType,thresh,alpha):
    init_time = time.time()  #记录开始事件
    i=0
    k=0 #当前批次的起始索引
    X,y = shuffleData(data)
    grad = np.zeros(theta.shape)
    costs=[cost(X,y,theta)]

    while True:
        #计算当前批次的梯度
        grad = gradient(X[k:k+batchSize],y[k:k+batchSize],theta)
        k+=batchSize
        if k >= n :
            k=0
            X,y = shuffleData(data)
        theta = theta - alpha*grad #沿着梯度反方向更新参数
        costs.append(cost(X,y,theta))
        i+=1

        if stopType == STOP_ITER : value = i
        elif stopType == STOP_COST: value = costs
        elif stopType == STOP_GRAD: value = grad

        if stopCriterion(stopType,value,thresh):break

    return theta,i-1,costs,grad,time.time() - init_time
```

```python
def runExpe(data,theta,batchSize,stopType,thresh,alpha):
    theta,iter,costs,grad,dur=descent(data,theta,batchSize,stopType,thresh,alpha)
    #判断函数是否经过了标准化
    name = "Original" if (data[:,1]>2).sum()>1 else "Scaled"   #判断是否为标准化数据
    name+="data - learning rate:{} - ".format(alpha) #在名称中添加学习率信息
    #根据批次大小区分梯度下降类型
    if batchSize == n : strDescType = "Gradient" #全量样本，批量梯度下降
    elif batchSize == 1 : strDescType="Stopchastic" #随机梯度下降
    else: strDescType = "Mini-batch({})".format(batchSize) #小批量
    name+=strDescType+"descent - Stop:"
    #根据停止准则补充名称
    if stopType == STOP_ITER : strStop="{} iterations".format(thresh) #按迭代次数停止
    elif stopType ==STOP_COST: strStop="costs change ，{}".format(thresh) #按代价变化停止
    else:strStop = "gradient norm<{}".format(thresh) #按梯度模长停止
    name+=strStop
    print("***{}\nTheta:{}-Iter:{}-Last cost :{:03.2f}-Duration:{:03.2f}s".format(name,theta,iter,costs[-1],dur))
    #可视化代价函数收敛曲线
    # 在创建 fig, ax = plt.subplots(...) 前添加
    plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']  # 优先使用支持中文的字体
    plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示异常问题（可选）
    fig,ax=plt.subplots(figsize=(12,4))
    ax.plot(np.arange(len(costs)),costs,'r')
    ax.set_xlabel('Iterations')
    ax.set_ylabel('Cost')
    ax.set_title(name.upper()+'-Error vs . Iteration')
    return theta
```

### 1.3精度

```python
def predict(X,theta):
    return [1 if x>=0.5 else 0 for x in model(X,theta)]
```

```python
#评估逻辑斯蒂回归模型的预测准确率
scaled_X = scaled_data[:,:3] #提取前三列特征
y=scaled_data[:,3]#提取真实标签
predictions=predict(scaled_X,theta)
correct=[1 if ((a==1 and b==1) or (a==0 and b==0)) else 0 for (a,b) in zip(predictions,y)]
accuracy = (sum(map(int,correct)) % len(correct))
print('accuracy={0}%'.format(accuracy))
```
