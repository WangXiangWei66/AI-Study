```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs  #生成可自定义的聚类数据
```

```python
data,target = make_blobs(centers=2)  #指定了两个聚类中心
plt.scatter(data[:,0],data[:,1],c = target)
```

```python
data.shape
```

```python
X = np.hstack((np.ones(shape=(100,1)),data))
```

```python
X.shape
```

```python
#预测函数
def sigmod(theta):
    z = np.dot(X,theta)
    return  1 / (1 + np.exp(-z))
```

```python
target
```

```python
np.log(0)
```

```python
#定义损失
def loss(y_true,y_pred):
    return -np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)) / 100
```

```python
#梯度下降求损失函数最小时候的theta
theta = np.random.randn(3) * 0.01
learning_rate = 0.01

max_iter = 20000
i = 0
while True:
    y_proba = sigmod(theta)  # 计算预测概率
    dtheta = np.dot((y_proba-target),X)/100
    theta = theta - learning_rate * dtheta

    i+=1
    if i % 100 == 0:
        y_proba = sigmod(theta)
        y_pred = (y_proba >= 0.5)
        acc = (y_pred == target).mean()
        loss_ = loss(target,y_proba)
        print(f'第{i}次训练，准确率：{acc},损失:{loss_}')

    if i > max_iter:
        break
```
