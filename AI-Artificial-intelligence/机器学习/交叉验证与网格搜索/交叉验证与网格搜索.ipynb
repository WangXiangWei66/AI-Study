```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
data = iris.data
target=iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,target,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
# 数据切割产生的影响, 造成模型评估不能客观的反应模型的性能. 
# 0.9 ,0.93,  0.9666, 1
```

```python
# 要想办法消除因为数据切割对算法性能的评估产生的不良影响. 
```

```python
#核心是将数据集分割为训练集和验证集
from sklearn.model_selection import KFold,StratifiedKFold
```

```python
kfold=KFold(n_splits=5,shuffle=True)
```

```python
scores=[]
for train,test in kfold.split(data,target):
    X_train=data[train]
    y_train=target[train]
    X_test=data[test]
    y_test=target[test]
    knn=KNeighborsClassifier()
    knn.fit(X_train,y_train)
    score=knn.score(X_test,y_test)
    scores.append(score)
    print(score)
```

```python
print(np.array(scores).mean())
```

```python
skf=StratifiedKFold(n_splits=5,shuffle=True)
```

```python
scores=[]
for train,test in skf.split(data,target):
    X_train=data[train]
    y_train=target[train]
    X_test=data[test]
    y_test=target[test]
    knn=KNeighborsClassifier()
    knn.fit(X_train,y_train)
    score=knn.score(X_test,y_test)
    scores.append(score)
    print(score)
```

```python
print(np.array(scores).mean())
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn = KNeighborsClassifier()
param_grid={
    'n_neighbors':range(2,20),
    'p':[1,2]  #p：距离度量参数，1 表示曼哈顿距离（L1），2 表示欧氏距离（L2）
}
#n_jobs=-1:表示使用CPU进行并行计算
gv = GridSearchCV(estimator=knn,param_grid=param_grid,n_jobs=-1,cv=5,verbose=3)
```

```python
gv.fit(data,target)
```

```python
gv.best_params_
```

```python
gv.best_estimator_
```

```python
gv.score(data,target)
```
