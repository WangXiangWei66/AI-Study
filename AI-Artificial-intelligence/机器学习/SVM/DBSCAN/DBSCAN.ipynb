```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN #用于无监督的聚类任务
from sklearn.metrics import silhouette_samples,silhouette_score
```

```python
data,target = make_circles(factor=0.2,noise=0.1)
plt.scatter(data[:,0],data[:,1],c=target)
plt.axis('equal')
```

```python
kmeans = KMeans(2)
kmeans.fit(data,target)
```

```python
plt.scatter(data[:,0],data[:,1],c=kmeans.labels_)
plt.axis('equal')
```

```python
# eps半径
# min_samples成一个簇所需要的最小样本数
# 一般先确定半径的范围. 然后再调min_samples, 最后微调半径. 
dbscan=DBSCAN(eps=0.37,min_samples=4)
dbscan.fit(data)
plt.scatter(data[:,0],data[:,1],c=dbscan.labels_)
plt.axis('equal')
```

```python
dbscan.labels_
```

```python
silhouette_score(data,dbscan.labels_)
```

```python
silhouette_samples(data, dbscan.labels_)
```
